##single inheritance
    A derived class inherits from a single parent class,gaining access its properties and methods.

```python
class Animal:
  def speak(self):
    print("Animal Speaking")
  #child class Dog inherits the base class animal
class Dog(Animal):
  def bark(self):
    print("dog barking")

d=Dog()
d.bark()
d.speak()
```

##Multilevel Inheritance:
    A class inherits from another derived class, forming a chain of inheritance

```python
class Animal:
  def speak(self):
    print("Animal Speaking")
  #child class Dog inherits the base class animal
class Dog(Animal):
  def bark(self):
    print("dog barking")
#The child clss dogchild inherits another child class Dog

class DogChild(Dog):
  def eat(self):
    print("Eating bread...")

d=DogChild()

d.bark()
d.speak()
d.eat()
```

```python
class Grandparent:
  def display(self):
    print("This is the grandparent class")
class parent (Grandparent):
  def message(self):
    print("This is the parent class")

class child(parent):
  pass

obj=child()
obj.display()
obj.message()
```

##Hierarchical Inheritance:
    Multiple derived classes inherit from a single parent class,allowing shared behavior across different subclasses

```python
class shape:
  def info(self):
    print("shapes have dimentions")
class Circle(shape):
  def draw(self):
    print("Drawing a circle")
class Rectangle(shape):
  def draw(self):
    print("Draw a Rectangle")

c=Circle()
r=Rectangle()
c.info()
c.draw()
r.info()
r.draw()
```

##Multiple  inheritance:
    A single derived class inherits from multiple parent classes,combining functionalities from different sourses

```python
class Calculation1:
  def Summation(self,a,b):

    return a+b
class Calculation2:
  def Multiplication(self,a,b):
    return a*b
class Derived(Calculation1,Calculation2):
  def Divide(self,a,b):

    return a/b
d=Derived()

print(d.Summation(10,20))
print(d.Multiplication(10,20))

print(d.Divide(10,20))
```

##Hybird Inheritence

```python
class person:
  def show(self):
    print("I am aperson")
class Teacher(person):
  def teach(self):
    print("I taech student")
class Staff:
  def manage(self):
    print("I manage school activity")

#Hybird inheritance: Principal inheritance from both Teacher and staff(single+Multiple)

class Principal(Teacher,Staff):
  def oversee(self):
    print("I oversee the entire school")

principal=Principal()
principal.show()
principal.teach()
principal.manage()
```

#Super()
    super() is a build-in function in python that allows a derived (child) class to access and invoke methods or constractors from its parent(base) class.It is commonly used to avoid code duplication and facilitate method resolution in multiple inheritance
      

      super().method_name(argument)

```python
class Parent:
  def greet1(self):
    print("Hello from parent")
class Child(Parent):
  def greet2(self):
    super().greet1() #calling the parent
    print("Hello From child")

child=Child()
child.greet2()
```

#Abstraction


*   Abstraction is a concept in Object-Oriented Programming (OOP) that focuss on hiding the implementation details and showing only the necessary feature of an object.
*   It allowa user to focus on what an object does rather than how it does it.

*   Example:
  *   When you use a smartphone,you only see the interface (buttons,icon),but you don't need to knowhow the internal circuits work.










##why is Abstraction important?


*   Simplifies complex system by focusing only on essential aspects

*   Reduses code complexity and increses reusability.
*   Enhances security by restricting direct access to implementation details





##Abstract in Python using Abstract classes

*   Python provides abstraction using abstract classes and abstract methods,Which are defined in the abc module

*   Abstract class: A class that contain one or more abstract method.It cannot be instantiated directly.

*   Abstract method: A method that is declared but contains no implementation (must be implemented by child classes).





How to implement Abstraction in python?


```python
from abc import ABC, abstractmethod

class Vehicle(ABC): #abstract class

  @abstractmethod
  def start_engine(self):
    pass #abstract method with no implementation


  @abstractmethod
  def stop_engine(self):
    pass #abstract method with no implementation


# making child class for abstract class instantiating

class car(Vehicle): # child class of vehicle

  def start_engine(self):  # also we need to create same function (method) in child class  other wise it will shows error!
    print("Car engine started")

  def stop_engine(self):   # also we need to create same function (method) in child class  other wise it will shows error!
    print("Car engine stopped")

car = car()
car.start_engine()
car.stop_engine()


#All child classes must implement the abstract method of the parent abstract class
#Abstract classes can have both abstract mrthod and concret(regular) methods
```

```python
class Shape(ABC):
  @abstractmethod
  def Calculate_area(self):
    pass
class Circle(Shape):
  def __init__(self,radius):
    self.radius=radius
  def  Calculate_area(self):
    return 3.14 * self.radius *self.radius
class Rectangle(Shape):
  def __init__(self,width,height):
    self.width=width
    self.height=height
  def Calculate_area(self):
    return self.width *self.height
cir=Circle(5)
rect=Rectangle(10,5)
cir.Calculate_area()
rect.Calculate_area()
```

Class method can modify class-level variable or properties (variables shared amoung all instance)This allows changes to be made that affect the class as a whole,rather than just an individual instance.

```python
class Car:
  #class vaeiable
  wheels=4

  def __init__(self,brand,model):
    self.brand=brand
    self.model=model

  @classmethod
  def change_wheels(cls,new_wheels):
    cls.wheels=new_wheels
    print(f'The number of wheel has been updated to {cls.wheels}')

car1=Car("Toyota","corolla")
Car.change_wheels(6)
print(car1.wheels)
```
